```python
plt.figure(figsize=(6,4))

for pid, traj in trajectories.items():
    if traj:
        xs, ys = zip(*traj)
        plt.plot(xs, ys, label=f"Player {pid}")

plt.gca().invert_yaxis()
plt.legend()
plt.title("Player trajectories")
plt.show()
```
